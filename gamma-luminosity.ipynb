```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
eng_mod = [x.split(' ')[0] for x in open("wind-gamma-murase-modest.txt").readlines()[1:]]
lum_mod = [x.split(' ')[1] for x in open("wind-gamma-murase-modest.txt").readlines()[1:]]

eng_mod = np.array(eng_mod, dtype=np.float64)
lum_mod = np.array(lum_mod, dtype=np.float64)
```

```python
eng_opt = [x.split(' ')[0] for x in open("wind-gamma-murase-optimistic.txt").readlines()[1:]]
lum_opt = [x.split(' ')[1] for x in open("wind-gamma-murase-optimistic.txt").readlines()[1:]]

eng_opt = np.array(eng_opt, dtype=np.float64)
lum_opt = np.array(lum_opt, dtype=np.float64)
```

```python
plt.plot(eng_mod, lum_mod, 'o', label='moderate wind')
plt.plot(eng_opt, lum_opt, 'o', label='optimistic wind')
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$\epsilon$ [GeV]')
plt.ylabel(r"$\epsilon L_{\epsilon}$ [erg/s]")


leg = plt.legend( loc='lower left')
```
